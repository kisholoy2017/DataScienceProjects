```python
import matplotlib.pyplot as plt
import numpy as np

from google.colab import files
uploaded = files.upload()

import pandas as pd
import io
 
```

```python
df = pd.read_excel(io.BytesIO(uploaded['category data.xlsx']))
print(df)
```

```python
grouped_df = df.groupby(['Month', 'Category']).sum()['Revenue'].reset_index()
```

```python
pivoted_df = grouped_df.pivot(index='Month', columns='Category', values='Revenue')
```

```python
ax = pivoted_df.plot(kind='bar', figsize=(12,8))
ax.set_xlabel("Month")
ax.set_ylabel("Revenue")
ax.set_title("Revenue by Product Category and Month")
plt.show()
```
